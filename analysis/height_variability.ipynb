```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/people.csv")
df
```

```python
def age_group(age):
  if age < 18:
    return "Child"
  elif age < 65:
    return "Adult"
  else:
    return "Senior"
  
df["age_group"] = df["age"].apply(age_group)
df
```

```python
df.boxplot(column="height", by="age_group")
plt.xlabel("Age Group")
plt.ylabel("Height (cm)")
plt.title("Height Variability by Age Group")
plt.suptitle("")
plt.show()
```

## Interpretation

The box plot shows that children have the widest spread in height, indicating high variability.
Adults have a more concentrated height distribution, while seniors show a lower median height with some variability.
This suggests that height variability differs meaningfully across age groups, not just in average values.
